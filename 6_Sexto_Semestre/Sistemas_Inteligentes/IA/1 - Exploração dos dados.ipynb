## Carregando a base de dados IRIS


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Carregando a base de dados Iris
iris = load_iris()
data = iris.data
target = iris.target

# Criando um DataFrame a partir dos dados
df = pd.DataFrame(data, columns=iris.feature_names)
df['target'] = target
# Mapear o target numérico para nomes das espécies para usar como hue em plots
try:
    df['species'] = [iris.target_names[t] for t in target]
    df['species'] = df['species'].astype('category')
except Exception:
    df['species'] = df['target'].astype(str)

# Mostrar o DataFrame
df
```

```python
df.describe()
```

```python
df.info()
```

```python
# Criar um boxplot para cada feature (usar melt para formatar os dados)
plt.figure(figsize=(10, 6))
df_melt = df.melt(id_vars=['species'], value_vars=iris.feature_names, var_name='feature', value_name='value')
sns.boxplot(data=df_melt, x='feature', y='value')
plt.title("Boxplots das Features no Conjunto Iris")
plt.xlabel("Features")
plt.ylabel("Valores")
plt.show()
```

```python
# Criando o gráfico usando o Seaborn
sns.pairplot(df, hue="species")
plt.suptitle("Gráfico das Características da Base de Dados Iris")
plt.show()
```
